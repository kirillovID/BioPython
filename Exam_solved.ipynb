Дана программа:

```
table = [{"gene": "CLIC4",   "val": 5.362225, "cond": True},
```

Выберите корректное описание того, что делает эта программа.

 - Создает множество `container`, в котором находятся удовлетворяющие условию `cond` гены, и выводит среднее арифметическое значение `val` для них.

 - Создает словарь `container`, в котором находится множество только удовлетворяющих условию `cond` генов, и выводит среднее арифметическое значение `val` для них.

 - Создает словарь `container`, в котором находятся множества как удовлетворяющих, так и не удовлетворяющих условию `cond` генов, и выводит среднее арифметическое значение `val` для первого из перечисленных множеств.

 - Создает словарь `container`, в котором находятся множества как удовлетворяющих, так и не удовлетворяющих условию `cond` генов, и выводит общее среднее арифметическое значение `val` для всех генов.

 - Создает множество `container`, в котором находятся как удовлетворяющие, так и не удовлетворяющие условию `cond` гены, и выводит среднее арифметическое значение `val` для них.

````python
# Создает словарь container, в котором находятся множества как удовлетворяющих,
# так и не удовлетворяющих условию cond генов,
# и выводит среднее арифметическое значение val для первого из перечисленных множеств.

table = [{"gene": "CLIC4",   "val": 5.362225, "cond": True},
         {"gene": "LAMA3",   "val": 3.495558, "cond": True},
         {"gene": "ENO1",    "val": 9.074930, "cond": False},
         {"gene": "S100A16", "val": 8.455659, "cond": True},
         {"gene": "ADAM15",  "val": 6.881728, "cond": False}]

var = 0
container = {}
for row in table:
    container[row["cond"]] = container.get(row["cond"], set()) | {row["gene"]}
    if row["cond"]:
        var += row["val"]
var /= len(container[True])
print(var)
````

 - Задан пустой список `gc_rich`. Этот код уже написан.
 - В цикле `while` считываются последовательности ДНК (каждая на своей строке) до тех пор, пока не введена пустая строка (не учитывается в последовательности).
 - Для каждой вводимой последовательности сделайте проверку: если `GC`-состав не меньше 50% — добавьте в список `gc_rich` кортеж из двух элементов: на первом месте последовательность, на втором — ее `GC`-состав, округленный до второго знака после запятой.
 - Гарантируется, что до пустой строки будет введена как минимум одна непустая.
 - Напечатайте список `gc_rich`.
 
```
gc_rich = []

# ваш код ниже
```

````python
gc_rich = []

# ваш код ниже
while seq := input():
    gc_content = (seq.count("G") + seq.count("C")) / len(seq)
    if gc_content >= 0.5:
        gc_rich.append((seq, round(gc_content, 2)))
print(gc_rich)
````

````python
gc_rich = []

# ваш код ниже
seq = input()
while seq:  # можно и while seq != "", но обычно так не пишут
    gc_content = 0
    for nucl in seq:
        if nucl in "GC":
            gc_content += 1
    gc_content /= len(seq)
    if gc_content >= 0.5:
        gc_rich.append((seq, round(gc_content, 2)))
    seq = input()
print(gc_rich)
````

 - Необходимо найти среди них те, экспрессия которых снизилась исключительно в результате первого, исключительно в результате второго или исключительно в результате третьего эксперимента. 
 - Полученные гены необходимо вывести в алфавитном порядке. Если же ни одного такого гена не оказалось, выведите строку "Каждый ген 'упал' в результате хотя бы двух экспериментов".

````python
first_genes = input().split()
second_genes = input().split()
third_genes = input().split()

# ваш код ниже
first_genes = set(first_genes)
second_genes = set(second_genes)
third_genes = set(third_genes)

overlap = first_genes & second_genes & third_genes
result = first_genes ^ second_genes ^ third_genes - overlap
if result:
    print(*sorted(result))
else:
    print("Каждый ген 'упал' в результате хотя бы двух экспериментов")
````

 - Записи вида `"<ДЕНЬ МЕСЯЦА> <СУММА>"`, состоящие из двух целых чисел, разделенных пробелом. Днем месяца может быть любое число от `1` до `31`, суммой может быть любое целое число. Каждая запись на новой строке. Дни месяца не повторяются, но могут идти не по порядку.
 - Гарантируется, что будет введена информация как минимум об одном дне.
 - Записи вводятся до тех пор, пока не будет введено число `0`.
 - После этого вводится целое число от `1` до `31` — верхняя граница интервала (включительно), для которого Алиса хочет найти общую сумму расходов.

**ФОРМАТ ВЫВОДА**

 - Одно целое число — общая сумма расходов Алисы в определенный интервал времени: от первого числа месяца до введенного числа включительно.
 - Если за этот период не было трат — выводится `0`.

````python
spent = [0] * 31
while (inp := input()) != "0":
    day_idx, amount = map(int, inp.split())
    spent[day_idx - 1] = amount
end = int(input())
print(sum(spent[:end]))
````

 - После финальной строки на вход подается индекс предмета в соответствующих списках оценок (длины всех списков оценок одинаковы).
 - Надо посчитать медиану оценок по соответствующему полученному индексу предмету.
 - Как считается медиана набора чисел: в упорядоченном наборе берется среднее арифметическое двух центральных элементов, если в наборе четное число элементов, и просто центральный элемент, если нечетное.
 
**ФОРМАТ ВЫВОДА**

 - В конце надо напечатать ответ: медианную оценку по предмету среди всех студентов. Округлите ответ до двух сотых.

````python
grades_dict = {}

while (name := input()) != "КОНЕЦ":
    grades_dict[name] = list(map(int, input().split()))

idx = int(input())
to_median = [grades[idx] for grades in grades_dict.values()]
to_median.sort()

left = (len(to_median) - 1) // 2
right = len(to_median) // 2 + 1
center_list = to_median[left:right]
print(round(sum(center_list) / len(center_list), 2))
````

````python
all_grades = []

while (name := input()) != "КОНЕЦ":
    one_student_grades = list(map(int, input().split()))
    all_grades.append(one_student_grades)
    
idx = int(input())
to_median = sorted([one_grades[idx] for one_grades in all_grades])

n = len(to_median)
if n % 2 == 0:
    print(round((to_median[n // 2 - 1] + to_median[n // 2]) / 2, 2))
else:
    print(to_median[n // 2])
````

**ФОРМАТ ВЫВОДА**

 - Словарь вида (просто вывести `print`-ом, необязательно так красиво):
```
{

        "жанр": ["научная фантастика", "драма"],
        "режиссер": "Дени Вильнёв",
        "актеры": ["Эми Адамс", "Джереми Реннер", "Форест Уитакер"],

    "Дюна": {

        "рейтинг": 8.0
    }

```

````python
movies = {}
while (movie := input()):
    info_dict = {}
    info = input()
    for one_info in info.split("; "):
        key, value = one_info.split(": ")
        if key == "режиссер":
            info_dict[key] = value
        elif key == "рейтинг":
            info_dict[key] = float(value)
        else:
            info_dict[key] = value.split(", ")
    movies[movie] = info_dict
print(movies)
````

## 5

Напишите программу, которая производит вычисление выражения, записанного в обратной польской нотации (ОПН).
В ОПН нет ни скобок, ни приоритета операторов («умножение раньше сложения»).

Чтобы прочитать выражение, записанное в таком формате, нужно просматривать выражение строго последовательно. Вводимые значения последовательно добавляются в стек (можно его себе представить как стопку карт внутри узкой коробки без крышки — вы можете видеть/снять только самую верхнюю карту, и положить карту можно только на стопку, внутрь замешивать нельзя). Когда встречается символ операции, то из стека извлекаются последние положенные туда значения, с ними проделывается эта операция и результат возвращается в стек.

Если для операции важен порядок значений, с которыми она производится, то первым идёт число, лежавшее в стеке глубже. В частности, если операция — вычитание, то из предпоследнего числа в стеке вычитается последнее, а не наоборот.

Изначально стек пустой, в результате полного вычисления выражения в нём должно остаться одно значение — результат вычислений. Сэмулировать стек вам поможет `list` — его методы `.pop()/.append()` идеально приспособлены для этого.

Операции, которые выполняются с одним значением, называются унарными, с двумя — бинарными, с тремя — тернарными. Давайте улучшим наш калькулятор, добавив поддержку следующих операций:

    бинарные:
        + (сложение),
        - (вычитание),
        * (умножение),
        / (деление нацело; для отрицательных чисел работает по тем же правилам, что и в Python);
    унарные:
        ~ (унарный минус — меняет знак),
        ! (факториал),
        # (клонирование — вернуть в стек значение дважды);
    тернарные:
        @ (возвращает в стек те же три значения, но в ином порядке: второе, третье, первое).

**ФОРМАТ ВВОДА**

Вводится одна строка, содержащая разделённые пробелами целые числа и знаки операций. Вместе они составляют корректное выражение в обратной польской нотации, не содержащее деления на ноль и взятия факториала от отрицательного числа.

**ФОРМАТ ВЫВОДА**

Выводится одно целое число — результат вычисления выражения.

**ПРИМЕЧАНИЕ**

В первом примере стек по мере прочтения строки выглядит так:

    7
    7 1
    7 1 10
    7 1 10 100
    7 1 10 100 100
    7 1 10 10000
    7 10 10000 1
    7 10 9999
    7 10009
    10016
    -10016

````python
stack = []
unary = {"~", "!", "#"}
binary = {"+", "-", "*", "/"}
ops = input().split()
for elem in ops:
    if elem.isnumeric():
        stack.append(int(elem))
    elif elem in unary:
        first = stack.pop()
        if elem == "~":
            stack.append(-first)
        elif elem == "!":
            res = 1
            for i in range(1, first + 1):
                res *= i
            stack.append(res)
        elif elem == "#":
            stack.extend([first, first])
    elif elem in binary:
        second, first = stack.pop(), stack.pop()
        if elem == "+":
            stack.append(first + second)
        elif elem == "-":
            stack.append(first - second)
        elif elem == "*":
            stack.append(first * second)
        elif elem == "/":
            stack.append(first // second)
    else:
        third, second, first = stack.pop(), stack.pop(), stack.pop()
        stack.extend([second, third, first])
print(stack.pop())
````
